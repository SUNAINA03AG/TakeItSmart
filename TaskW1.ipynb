```python
# CLASS 2 â€“ Rock Paper Scissors

import random

print("Welcome to Rock Paper Scissors")

while True:
    user = input("Type rock/paper/scissors (or exit): ")

    if user == "exit":
        break

    if user not in ["rock", "paper", "scissors"]:
        print("Invalid input. Try again.")
        continue

    computer = random.choice(["rock", "paper", "scissors"])

    print("You chose:", user)
    print("Computer chose:", computer)

    if user == computer:
        print("Draw")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("You win")
    else:
        print("Computer wins")
```

```python

def monkey_trouble(a_smile, b_smile):
    return a_smile == b_smile


def sum_double(a, b):
    if a == b:
        return (a + b) * 2
    return a + b


def diff21(n):
    if n > 21:
        return (n - 21) * 2
    return 21 - n


def makes10(a, b):
    return a == 10 or b == 10 or (a + b) == 10


def near_hundred(n):
    return abs(n - 100) <= 10 or abs(n - 200) <= 10


def not_string(s):
    if s.startswith("not"):
        return s
    return "not " + s


def missing_char(s, n):
    return s[:n] + s[n+1:]
```

```python
# Exchange first and last characters
def front_back(s):
    if len(s) <= 1:
        return s
    return s[-1] + s[1:-1] + s[0]


# Repeat first 3 characters 3 times
def front3(s):
    return s[:3] * 3
```

```python
my_list = [2, 3, 4, 5, 6, 7, 8]
print(my_list)

# Add element
new_value = int(input("Enter value to add: "))
my_list.append(new_value)
print("Updated list:", my_list)

# Remove element
remove_item = int(input("Enter value to remove: "))
if remove_item in my_list:
    my_list.remove(remove_item)
    print("Updated list:", my_list)
else:
    print("Value not found")

# Search element
search = int(input("Enter value to search: "))
if search in my_list:
    print("Value found")
else:
    print("Value not found")
```

```python
my_list = []

print("Welcome to List Manager")
print("1.ADD\n2.REMOVE\n3.REPLACE\n4.VIEW\n5.EXIT")

while True:
    choice = int(input("ENTER UR CHOICE"))

    if choice == 1:
        count = int(input("How many elements? "))
        for i in range(count):
            value = int(input("Enter value: "))
            my_list.append(value)
        print("Updated list:", my_list)

    elif choice == 2:
        value = int(input("Enter value to remove: "))
        if value in my_list:
            my_list.remove(value)
            print("Updated list:", my_list)
        else:
            print("Value not found")

    elif choice == 3:
        old = int(input("Enter old value: "))
        if old in my_list:
            new = int(input("Enter new value: "))
            index = my_list.index(old)
            my_list[index] = new
            print("Updated list:", my_list)
        else:
            print("Value not found")

    elif choice == 4:
        print("Current list:", my_list)

    elif choice == 5:
        print("Program ended")
        break

    else:
        print("Invalid option")
```

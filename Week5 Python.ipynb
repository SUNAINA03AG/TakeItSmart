```python
 # Monday march2 class21

#we worked on #student_performance.csv dataset

#clean the data and see if u have any null or duplicate values

import pandas as pd
import numpy as np
df=pd.read_csv("student.csv")

print("dataset shape:",df.shape)
print(df.head())

df=df.drop(columns=["StudentID","Name"])

print("\n Missing values per column:")
print(df.isnull().sum())

duplicates=df.duplicated().sum()
print("\n Duplicate Rows:",duplicates)

df=df.drop_duplicates()

print("Shape After Removing Duplicates:",df.shape)

#fiiling the missing values
for i in df.columns:
  if df[i].dtype in ['int64','float64']:
    med_val=df[i].median()
    df[i]=df[i].fillna(med_val)
  else:
    mode_val=df[i].mode()[0]
    df[i]=df[i].fillna(mode_val)

print("\n Missing values per column:")
print(df.isnull().sum())

#gen data to numerical data ----

#Gender -- Binary Encoding
df["Gender"]=df["Gender"].map({"Male":0,"Female":1})

df["Online classes Taken"]=df["Online Classes Taken"].astype(int)

df["ParentalSupport"]= df["ParentalSupport"].map({
    "Low":0,
    "Medium":1,
    "High":2
})


def categorize(score):
  if score < 60:
    return 0   #Low
  elif score <80:
    return 1  #Medium
  else:
    return 2  #High

df["Performance_Level"]= df["FinalGrade"].apply(categorize)

#Train Test Split
#underfitting --

#Train Test Split accuraccy and matrix confusion


```

```python
#class24 Saturday

labelled data: data which has bot input and output
unlabelled data: data which has only input
eg:basketful of fruits

 supervised learning: it requires labelled data
                      it requires both input and output
                      eg:house price prediction
                        Diabetic prediction

 types of supervised learning:
 1.Regression
 2.classification

 Continuous data: Numerical data which lies within certain range
                 eg: heights,weight,price,sales,age.

 Categorical data: text data/categories
                   eg:diabetic/non diabetic
                     yes/no
                     cancer/non cancerous

 Regression:comes under supervised learning
            Regression is used when datatype of target variable is continuous
            Regression is used to predict continuous target variable
            eg: sales prediction
                house price prediction
                student marks prediction
                car price prediction
                flight price prediction

  Algorithm comes under regression:
  1.linear regression
  2.KNN regression
  3.SVM regression
  4.Decision tree regression
  5.Random Forest regression
  6.XGBoost regression

  Classification:comes under supervised learning.
                 it is used when datatype of target variable is
                 categorical/discrete
                 it is used to predict categories/classes.
                 eg:Diabetic prediction
                    cancer prediction
                    heart attack risk

  Algorithms of classification:
  1.logistic regression
  2.KNN classifier
  3.SVM classifier
  4.Decision tree classifier
  5.Random Forest classifier
  6.XGBoost classifier
  7.gradient boosting

  Unsupervised learning: It requires unlabelled data.
                         it requires only input
                         also known as clustering algorithmwhich takes unlabelled data and creates clusters/group based
                         on similarities
                         eg: market basket analysis

  Reinforcement learning: Learning happens based on rewards and punishments
                          It will interact with environment which focus on maximising the rewards
                          eg: self driving cars

  Semi supervised learning: Requires both labelled as well as unlabelled data
                            Model will be trained on small amount of labelled data and that model will be used to  make predictions related to unlabelled data
```

```python
#palindrome

def check_palindrome(data):
    if isinstance(data, int):
        data = str(data)
    return data == data[::-1]

print(check_palindrome(121))
print(check_palindrome("madam"))
print(check_palindrome([1,2,1]))
```

```python

A = [[1,2,3],[4,5,6],[7,8,9]]
B = [[9,8,7],[6,5,4],[3,2,1]]

result = [[0]*3 for _ in range(3)]

for i in range(3):
    for j in range(3):
        for k in range(3):
            result[i][j] += A[i][k] * B[k][j]

for row in result:
    print(row)
```

```python

employees = {
    "Aman":[30000]*12,
    "Riya":[50000]*12,
    "Sam":[20000]*12
}

highest = 0
top_emp = ""

for name, sal in employees.items():
    yearly = sum(sal)
    print(name, "Yearly:", yearly)

    if yearly > highest:
        highest = yearly
        top_emp = name

    if yearly < 300000:
        print(name, "earns < 3 lakh")

    if yearly > 500000:
        print(name, "gets bonus:", yearly*0.1)

print("Highest Earning:", top_emp)
```

```python

library = {"Python":5, "Java":2}

while True:
    print("1.Borrow 2.Return 3.Add 4.Show 5.Exit")
    ch = input()

    if ch == "1":
        book = input("Book: ")
        if library.get(book,0) > 0:
            library[book] -= 1
        else:
            print("Out of stock")

    elif ch == "2":
        book = input("Book: ")
        library[book] += 1

    elif ch == "3":
        book = input("Book: ")
        qty = int(input("Qty: "))
        library[book] = qty

    elif ch == "4":
        print(library)

    elif ch == "5":
        break
```

```python
import string

def check_password(pwd):
    score = 0
    if len(pwd) >= 8:
        score += 1
    if any(c.isupper() for c in pwd):
        score += 1
    if any(c.islower() for c in pwd):
        score += 1
    if any(c.isdigit() for c in pwd):
        score += 1
    if any(c in string.punctuation for c in pwd):
        score += 1

    if score == 5:
        return "Strong"
    elif score >= 3:
        return "Medium"
    else:
        return "Weak"

print(check_password("Abc@1234"))
```

```python

lst = [1,1,1,2,2,3,4,4,4,4]
result = []

count = 1
for i in range(1,len(lst)):
    if lst[i] == lst[i-1]:
        count += 1
    else:
        result.append((lst[i-1],count))
        count = 1

result.append((lst[-1],count))
print(result)
```

```python

votes = {}
voters = set()

while True:
    voter_id = input("Voter ID (or exit): ")
    if voter_id == "exit":
        break

    if voter_id in voters:
        print("Already voted")
        continue

    candidate = input("Candidate: ")
    votes[candidate] = votes.get(candidate,0) + 1
    voters.add(voter_id)

max_votes = max(votes.values())
winners = [k for k,v in votes.items() if v==max_votes]

if len(winners) > 1:
    print("Tie between", winners)
else:
    print("Winner:", winners[0])
```

```python

# 10. Sudoku Row Validator
def validate_sudoku(board):
    for row in board:
        if sorted(row) != list(range(1,10)):
            return "Invalid"
    return "Valid"
```

```python

#12. Anagram Grouping
words = ["eat","tea","tan","ate","nat","bat"]
result = {}

for word in words:
    key = "".join(sorted(word))
    result.setdefault(key,[]).append(word)

print(result)
```

```python

#13. Attendance Tracker
attendance = {
    "Aman":["P","A","P","P"],
    "Riya":["P","P","P","P"]
}

highest = 0
regular = ""

for name, record in attendance.items():
    percent = record.count("P")/len(record)*100
    print(name, percent,"%")

    if percent < 75:
        print(name,"Detained")

    if percent > highest:
        highest = percent
        regular = name

print("Most Regular:", regular)
```

```python

# 14. Prime Pattern Pyramid
def is_prime(n):
    if n < 2:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True

num = 2
for i in range(1,6):
    count = 0
    while count < i:
        if is_prime(num):
            print(num,end=" ")
            count += 1
        num += 1
    print()
```

```python

# 15. Multi-Level Inventory System

inventory = {
    "Electronics":{"Mobile":5,"Laptop":2},
    "Clothes":{"Shirt":10,"Jeans":4}
}

def search_product(prod):
    for cat,items in inventory.items():
        if prod in items:
            print("Found in",cat,"Qty:",items[prod])

search_product("Mobile")
```

```python
#16. Word Frequency + Longest Word
sentence = input("Enter sentence: ").split()
freq = {}

for word in sentence:
    freq[word] = freq.get(word,0)+1

longest = max(sentence, key=len)
most = max(freq, key=freq.get)

print("Longest:", longest)
print("Most Repeated:", most)
```

```python
#17. Nested List Sorting Without sort()
data = [[1,5],[2,3],[4,1]]

for i in range(len(data)):
    for j in range(i+1,len(data)):
        if data[i][1] > data[j][1]:
            data[i],data[j] = data[j],data[i]

print(data)
```

```python

#18. Exam Result Dashboard
classes = {
    "ClassA":{"Aman":80,"Riya":95,"Sam":35},
    "ClassB":{"John":78,"Anna":88}
}

overall_top = ("",0)

for cname, students in classes.items():
    topper = max(students, key=students.get)
    avg = sum(students.values())/len(students)
    print(cname,"Topper:",topper,"Average:",avg)

    for name,mark in students.items():
        if mark < 40:
            print(name,"Failed")

        if mark > overall_top[1]:
            overall_top = (name,mark)

print("Overall Topper:", overall_top[0])
```
